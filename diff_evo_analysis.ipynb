```python
import pickle 
import sys
import pandas as pd
import os
from plotnine import *
%matplotlib inline
```

```python
optimizers = []
dirname = 'log'
for dirpath, _, filenames in os.walk(dirname):
    for filename in filenames:
        if filename.endswith('.pkl'):
            with open(os.path.join(dirpath, filename), 'rb') as fp:
                optimizers.append(pickle.load(fp))
len(optimizers)
```

```python
import collections

def flatten(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, collections.MutableMapping):
            items.extend(flatten(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)

def max_profit(optimizer):
    return max(optimizer._population, key=lambda x: x.profit).profit
```

```python
df = pd.DataFrame(columns=['convergence_criterion_n_max_iterations',
                           'crossover_crossover_rate',
                           'differential mutation_scaling factor',
                           'population size', 
                           'profit'])

for i in range(len(optimizers)):
    df = df.append(dict(**flatten(optimizers[i].parameters), profit=max_profit(optimizers[i])), ignore_index=True)
df
```

```python
df['cat_differential_mutation_scaling_factor'] = df['differential mutation_scaling factor'].astype('category')
df['cat_crossover_rate'] = df['crossover_crossover_rate'].astype('category')

df
```

```python
import plotnine
plotnine.options.figure_size = (6.4, 4.8)
(ggplot(df, aes('cat_crossover_rate', 'profit'))
 + geom_jitter(width=0.1)
 + geom_boxplot(aes(color='cat_differential_mutation_scaling_factor')) 
)
```
